Llamamos **algoritmo** a un conjunto finito de instrucciones dadas en una sucesin de operaciones para resolver algn tipo especfico de problema. Caractersticas:   

1. **Finito:** terminar en una cantidad finita de pasos.
2. **Bien definido:** cada paso debe ser dado de forma *precisa* (lenguaje de programacin).
3. **Recibe entrada:** recibe cero o ms datos de entrada.
4. **Arroja salida:** produce uno o ms resultados a partir de los datos de entrada.

## <span style="color:#8B0000;font-family:Papirus">Ejemplo</span>

Vamos a describir el algoritmo de Euclides para obtener el mximo divisor comn $d$ de dos nmeros enteros $a$, $b$.

1. Divide $a$ entre $b$ y llama $r$ al residuo.
2. Si $r$ es cero entonces $d$ es $b$. Terminamos.
3. Sustituye $a$ por $b$ y $b$ por $r$ ($a\leftarrow b$, $b\leftarrow r$). Vuelve al paso 1.

La siguiente tabla contiene los valores en cada paso para $a = 119$ y $b = 544$: 

| $a$  |  $b$  |  $r$  |  
|:----:|:-----:|:-----:|
|$119$ | $544$ | $119$ |
|$544$ | $119$ |  $68$ |
|$119$ |  $68$ |  $51$ |
| $68$ |  $51$ |  $17$ |
| $51$ |  $17$ |   $0$ |

El MDC de $119$ y $544$ es $17$ (segn Euclides).

Escribiremos los algoritmos como **funciones** de python. Por ejemplo, el algoritmo de Euclides se puede escribir as:

```python
def MDC(a, b):
    r = a % b
    while r != 0:
        a = b
        b = r
        r = a % b
    return b
```

```python
MDC(5,35)
```

```python
MDC(3456, 23649)
```

## <span style="color:#8B0000;font-family:Papirus">Algoritmos "correctos"</span>

Recuerda que para que una lista de instrucciones sea un *algoritmo*, tiene que ser finito. Por ejemplo:

```python
def funcion(n):
    while n > 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 5 * n + 1
    return n
```

```python
funcion(3)
```

Pero si revisamos los valores producidos al tratar de calcular ``funcion(5)``:

| $n$  | 
|:----:|
|  $5$ |

| $66$ |
| $33$ |
|$166$ |
|$83$  |
|$416$ |

|$104$ |
| $52$ |

Se repite indefinidamente. No termina el proceso. En cambio, el algoritmo de Euclides s termina. En general, el proceso es:

$$a = b\cdot q_1 + r_1, \qquad (0 \leq r_1 < b)$$
$$b = r_1\cdot q_2 + r_2, \qquad (0 \leq r_2 < r_1)$$
$$r_1 = r_2\cdot q_3 + r_3, \qquad (0 \leq r_3 < r_2)$$
$$r_2 = r_3\cdot q_4 + r_4, \qquad (0 \leq r_4 < r_3)$$
$$ \vdots \qquad \qquad \vdots $$
$$r_{n - 4} = r_{n - 3}\cdot q_{n - 2} + r_{n - 2}, \qquad (0 \leq r_{n - 2} < r_{n - 3})$$
$$r_{n - 3} = r_{n - 2}\cdot q_{n - 1} + r_{n - 1}, \qquad (0 \leq r_{n - 1} < r_{n - 2})$$
$$r_{n - 2} = r_{n - 1}\cdot q_n, \qquad r_n = 0$$ 

El proceso se detiene en el paso $n$ ya que $r_n = 0$. Y esto ocurre independientemente de $a$ y $b$ ya que la sucesin $\{r_j\}_j$ es decreciente; y como se trata de enteros positivos, eventualmente se llega a cero. Este razonamiento se basa en el **principio de buen orden** en $\mathbb{N}$:

> Todo conjunto no vaco de nmeros naturales tiene un mnimo.

<span style="color:#8B0000;font-family:Papirus">**Corolario:**</span> Toda sucesin *estrictamente* decreciente de nmeros naturales es finita.

Muy bien, es un algoritmo porque tiene las caractersticas necesarias. Pero, cmo sabemos que en efecto calcula el MCD de los nmeros dados? Veamos, segn Euclides, el MCD de $a$ y $b$ es $r_{n - 1}$.

1. $r_{n - 1}$ divide a $a$ y a $b$? Por la ltima ecuacin, $r_{n - 1}$ divide a $r_{n - 2}$.
Y por la penltima ecuacin, divide a $r_{n - 3}$. Siguiendo de esta manera, vemos que divide a $b$ y finalmente, divide a $a$.
2. Supongamos que $d$ divide a $a$ y a $b$. Entonces, por la primera ecuacin, $d$ tambin divide a $r_1$. Y por la segunda, $d$ divide a $r_2$. Siguiendo as, vemos que divide a $r_{n - 1}$.

 Por 1 y 2, $MDC(a, b) = r_{n - 1}$.

<span style="color:#8B0000;font-family:Papirus">**Otro ejemplo:**</span> Veamos que el siguiente proceso termina (es un algoritmo) y descubramos qu es lo que calcula, dados los nmeros naturales de entrada $n$ y $m$.

```python
def S(n, m):
    while m > 0:
        n = n + 1
        m = m - 1
    return n
```

Fjate que la prueba para el *mientras* es el valor de $m$. Como ste va disminuyendo, eventualmente ser negativo. Por el principio del buen orden, el proceso terminar. Qu es lo que calcula este algoritmo? Probemos con varios valores de $n$ y $m$.

```python
S(10, 20)
```

Parece que es la suma de $n$ y $m$. Veamos: digamos que en el paso $j$, los valores de $n$ y $m$ son $n_j$ y $m_j$ respectivamente. Si $m_j > 0$ cambian los valores:
$$n_{j + 1} = n_j + 1, \qquad m_{j + 1} = m_j - 1$$
Cul es la suma de estos nuevos valores? 
$$n_{j + 1} + m_{j + 1} = n_j + 1 + m_j - 1 = n_j + m_j$$
<span style="color:#8B0000;font-family:Papirus">**Conclusin:**</span> la suma de los valores *no cambia* con los pasos. En este sentido, decimos que $n + m$ es **invariante** por el algoritmo. El resultado que arroja el algoritmo es el ltimo de los $n_j$. Veamos:
$$n_0 = n\qquad m_0 = m$$
$$n_1 = n + 1\qquad m_1 = m - 1$$
$$n_2 = n + 2\qquad m_2 = m - 2$$
$$\vdots$$
$$n_{m - 1} = n + m - 1\qquad m_{m - 1} = m - (m - 1) = 1$$
$$n_m = n + m\qquad m_m = 1 - 1 = 0$$

Queda demostrado que $S(n, m)$ calcula la suma de $n$ y $m$.
